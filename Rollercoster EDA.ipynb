```python
df1["opening_date_clean"]=pd.to_datetime(df1['opening_date_clean'])
```

```python
#rename some of ou columns
df1=df1.rename(columns={"coaster_name":"Coaster_Name", "year_introduced":"Year_Introduced", 
                    "opening_date":"Opening Date", "speed_mph":"Speed_mph",
                   "Inversion_clean":"Inversion_Clean", "Gforce_clean":"Gforce_Clean"})
df1
```

```python
df1.isna().sum()
#Any column that doesnt have a 0, have that many number of null values
```

```python
#Lets check for duplicate values
#this function always look for a value that is similar to its previous row.
df1.duplicated().sum()
#its good that we dont have any duplicates here
```

```python
df1['Coaster_Name'].duplicated().sum()
#The coaster name column has 97 duplicates
```

```python
#we can find the duplicates using the subset method
df1.loc[df1.duplicated(subset = ["Coaster_Name","Location","opening_date_clean"])].head(5)
# Its shows the duplicate rows the second time they occur
```

```python
df1=df1.loc[~df1.duplicated(subset=['Coaster_Name','Location', 'opening_date_clean'])]\
            .reset_index(drop=True)

#so we created a subset of the not duplicated values by using "~" and then created a new dataframe by reseting the indexes.
# so the duplicated values are eliminated now.
```

```python
#So the question is:
# What are the locations with the fastest roller coasters with a minimum of 10 coasters at that location?
df1['Location'].value_counts()
#sns.histplot(data=df1,x='Location', y='Speed_mph')
```

```python
dx=df1.query('Location != "Other"')\
    .groupby('Location')['Speed_mph']\
    .agg(['mean', 'count'])\
    .query('count>=10')\
    .sort_values('mean')\
    .plot(kind = 'bar', figsize = (12,5), title = 'Average coaster speed by location')

dx.set_xlabel('Average Coaster Speed')
dx.set_ylabel('Location')
plt.show
```
