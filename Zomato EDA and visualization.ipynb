```python
#lets clean the rate columns. outof the 4.1/5 we only need the 4.1
#first lets check the unique values in the rate column
df['rate'].unique()
```

```python
#lets deal with the 'Rate column'. First we create a custom function and then apply that function to the column
def handlerate(value):
    if (value=='NEW' or value=='-'):
        return np.nan
    else:
        value = str(value).split('/')
        value = value[0]
        return float(value)

```

```python
df['rate']=df['rate'].apply(handlerate)
df['rate'].head()
```

```python
df.head()
```

```python
#filling null values in the rate column with mean
df.rate.isnull().sum()

#we have almost 10000 null values and we are not going to remove the null values.
#instead we are going to replace those null values with the mean of all the rate values
```

```python
df['rate'].fillna(df['rate'].mean(),inplace=True)
```

```python
df.rate.isnull().sum()
#al the 10019 nan values have been replaced by the mean of the rate column
```

```python
df.shape
# so we have 51609 rows in total. Lets check for null values in other columns
```

```python
df.head()
# for location, rest_type, cuisines we can drop the null values as they are categorical variables
# for  approx cost for two people we can either drop the null values or replace them with the mean
```

```python
df.dropna(inplace=True)
```

```python
#lets see the unique values in cost2plate
```

```python
df['cost2plates'].unique()
# note how some values contain a comma. we need to remove those 
```

```python
#lets create a function to handle the comma
def handlecomma(value):
    value = str(value)
    if ',' in value:
        value = value.replace(',','')
        return float(value)
    else:
        return float(value)
    
df['cost2plates']=df['cost2plates'].apply(handlecomma)
df['cost2plates'].unique()
```

```python
#lets clean rest_type column
df['rest_type'].unique()
len(df['rest_type'].unique())
```

```python
rest_types=df['rest_type'].value_counts()
rest_types
```

```python
# we are going to make list of any type that is greater than 1000
rest_types_lessthan1000 = rest_types[rest_types<1000]
rest_types_lessthan1000
#now lets convert these types to another type as 'others'
#for that we write a function
```

```python
def handle_rest_type(value):
    if value in rest_types_lessthan1000:
        return 'others'
    else:
        return value

    
df['rest_type']= df['rest_type'].apply(handle_rest_type)
df['rest_type'].value_counts()
```

```python
df.head()
```

```python
#Lets clean the location columns
```

```python
location_count=df['location'].value_counts()
location_count_lessthan300 = location_count[location_count<300]
location_count_lessthan300
# so any location that is less than 300, we are going to label it as others
```

```python
def handle_location(value):
    if value in location_count_lessthan300:
        return 'others'
    else:
        return value

    
df['location']= df['location'].apply(handle_location)
df['location'].value_counts()
```

```python
df.head()
```

```python
df.shape
```

```python
#Now lets clean the cuisines column
df['cuisines'].unique()
```

```python
cuisine_count=df['cuisines'].value_counts(ascending=False)
cuisine_count
```

```python
cuisines_lessthan100=cuisine_count[cuisine_count<100]
cuisines_lessthan100
```

```python
def handle_cuisines(value):
    if (value in cuisines_lessthan100):
        return 'others'
    else:
        return value
```

```python
df['cuisines']=df['cuisines'].apply(handle_cuisines)
df['cuisines'].value_counts()
```
